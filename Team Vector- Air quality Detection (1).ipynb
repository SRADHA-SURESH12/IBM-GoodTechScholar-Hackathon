```python
import numpy as np 
import pandas as pd 
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("newhour.csv")
data
```

```python
data['Datetime']=pd.to_datetime(data['Datetime'])
```

```python
import missingno as msno
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

missing=pd.DataFrame(data.isna().sum()/len(data))
missing.columns=['Proportion']

msno.bar(data)
```

```python
data.fillna(0,inplace=True)

data['date'] = pd.DatetimeIndex(data['Datetime']).date
data['time'] = pd.DatetimeIndex(data['Datetime']).time


data['PM']=data['PM2.5'] + data['PM10']


data['Nitric']=data['NO'] + data['NO2']+ data['NOx']


data['BTX']=data['Benzene'] + data['Toluene']+ data['Xylene']


datagrp=data.groupby(['City','date','time'])[['PM','Nitric','CO','NH3','O3','SO2','BTX','AQI']].mean()

datagrp=datagrp.reset_index(['City','date','time'])
datagrp.head()
```

```python
for i in datagrp:
    if i=="AQI":
        for j in datagrp["AQI"]:
            if j>0 and j<50:
                print("Good")
            elif j>51 and j<100:
                print("Satisfactory")
            elif j>101 and j<200:
                print("moderately polluted")
            elif j>201 and j<300:
                print("poor")
            elif j>301 and j<400:
                print("very poor")
            else:
                print("severe")
```

```python
dftest = adfuller(data['AQI'], autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['statiscal result','p value','no of lags','observation count'])
for key,value in dftest[4].items():
    dfoutput['critical value (%s)'%key] = value
dfoutput
```

```python
trainset=data['AQI'][:2018]
testset=data['AQI'][:2019]
model=SARIMAX(trainset,order=(1,1,0),seasonal_order=(2,0,[1,2],12), trend='n')
final=model.fit()
final.summary()
```

```python
fig, ax= plt.subplots(figsize=(10,6))
forecastdata= final.get_forecast(steps=12, dynamic=True)
conf_int=forecastdata.conf_int()
low = conf_int.loc[:,'lower AQI']
upper = conf_int.loc[:,'upper AQI']
forecastdata.predicted_mean.plot(legend=True, ax=ax, label ='Predicted AQI')
plt.fill_between(conf_int.index, low, upper, color='green')

testset.plot(legend=True, ax=ax)
```
